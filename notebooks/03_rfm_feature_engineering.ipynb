```python
snapshot_date = df['InvoiceDate'].max()
```

```python
invoice_totals = df.groupby(['CustomerID', 'InvoiceNo'])['TotalPrice'].sum().reset_index()

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days
}).reset_index()

rfm.columns = ['CustomerID', 'Recency']
```

```python
freq_mon = invoice_totals.groupby('CustomerID').agg({
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
}).reset_index()

freq_mon.columns = ['CustomerID', 'Frequency', 'Monetary']

rfm = rfm.merge(freq_mon, on='CustomerID')
rfm.head()
```
